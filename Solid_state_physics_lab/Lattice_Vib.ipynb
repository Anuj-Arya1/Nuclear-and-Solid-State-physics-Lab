```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Monoatomic lattice parameters: f_c = 26.36 kHz, L = 0.97 mH, C = 150.3 nF, N = 10 units

fm = [0.025,1.705,3.85,5.9,7.95,9.85,11.9,13.8,15.65,17.3,19,20.55,22,23.25,24.55,25.75,26.8] # in kHz
phase_m = [0,90,180,270,360,450,540,630,720,810,900,990,1080,1170,1260,1350,1440] # in degrees
phase_10 = np.array(phase_m)/10 # phase shift per unit

plt.plot(phase_10,fm, marker='*', markerfacecolor = 'red',linestyle='--', color='black', label='Monoatomic Lattice')
plt.title('Monoatomic Lattice Vibrations: Phase Shift per Unit Cell vs Frequency')
plt.xlabel('Frequency (kHz)')
plt.ylabel('Phase Shift per unit cell(degrees)')
plt.legend()
plt.grid()
plt.show()

# Diatomic lattice parameters: f_c = 26.36 kHz, L = 0.97 mH, C1 = 150.3 nF,C2 = 46.64 nF, N = (5 units

fd = [0.025,2.105,4.55,7.05,9.45,11.7,13.95,15.95,17.8,19.85,33.5,35.5,36.5,37.5,38.5,39.5] # in kHz
phase_d = [0,90,180,270,360,450,540,630,720,810,900,990,1080,1170,1260,1350] # in degrees
phase_5 = np.array(phase_d)/5 # phase shift per unit

fd_plot = np.array(fd)
fd_plot = np.insert(fd_plot, 10, np.nan)
phase_plot = np.insert(phase_5, 10, np.nan)
# mark acoustic and optical branches with different colors
plt.plot(phase_plot[:10],fd_plot[:10], marker='*', markerfacecolor = 'red',linestyle='--', color='black', label='Acoustic Branch')
plt.plot(phase_plot[10:],fd_plot[10:], marker='>', markerfacecolor = 'red',linestyle='--', color='black', label='Optical Branch')  
plt.title('Diatomic Lattice Vibrations: Phase Shift per Unit Cell vs Frequency')    
plt.xlabel('Frequency (kHz)')
plt.ylabel('Phase Shift per unit cell(degrees)')
plt.legend()
plt.grid()
plt.show()
```
