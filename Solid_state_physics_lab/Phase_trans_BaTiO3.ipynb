```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# frequency dependent data for BaTiO3
C0 =  (8.85*10**(-3)*(np.pi*(10.56/2)**2))/1.6 # pF
f = [10,15,20,25,30,35,40,45,50] # kHz
C4 = [350,300,250,200,200,150,150,150,150] # pF
R4 = [0.884,0.878,0.876,0.872,0.870,0.868,0.866,0.864,0.864] # kOhm
C1 = [(1000*R4[i])/3 for i in range(len(C4))] # pF
R1 = [ (3*C4[i])/1000 for i in range(len(C4))] # kOhm
e = [C1[i]/C0 for i in range(len(C4))] # dimensionless
D_fac = [2*np.pi*f[i]*C1[i]*R1[i]*10**(-6) for i in range(len(C4))] # dimensionless

print("Frequency (kHz) | C1 (pF) | R1 (kOhm)| e      | D_fac")
for i in range(len(C4)):
    print(f"             {f[i]} | {C1[i]:.3f}  | {R1[i]:.3f}     | {e[i]:.3f} | {D_fac[i]:.3f}")

# Plotting
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(f, e, marker='o')
plt.title('Dielectric Constant (e) vs Frequency')
plt.xlabel('Frequency (kHz)')
plt.ylabel('Dielectric Constant (e)')
plt.subplot(1, 2, 2)
plt.plot(f, D_fac, marker='o', color='orange')
plt.title('Dissipation Factor (D_fac) vs Frequency')
plt.xlabel('Frequency (kHz)')
plt.ylabel('Dissipation Factor (D_fac)')
plt.tight_layout()
plt.show()
```

```python
# for MLCC and disc ceramic capacitors
f_mlcc = [10,15,20,25,30,35,40,45,50] # kHz
R4_mlcc = [0.592,0.590,0.588,0.588,0.588,0.588,0.586,0.588,0.588] # kOhm 
C1_mlcc = [ (1000*R4_mlcc[i])/3 for i in range(len(R4_mlcc))] # pF

f_disc = [10,15,20,25,30,35,40,45,50] # kHz
R4_disc = [0.528,0.528,0.526,0.526,0.526,0.524,0.524,0.524,0.522] # kOhm
C1_disc = [ (1000*R4_disc[i])/3 for i in range(len(R4_disc))] # pF

print("\nMLCC Capacitors:")
print("Frequency (kHz) | R4 (kOhm) | C1 (pF)")
for i in range(len(f_mlcc)):
    print(f"             {f_mlcc[i]} | {R4_mlcc[i]:.3f}     | {C1_mlcc[i]:.3f}")

print("\nDisc Ceramic Capacitors:")
print("Frequency (kHz) | R4 (kOhm) | C1 (pF)")
for i in range(len(f_disc)):
    print(f"             {f_disc[i]} | {R4_disc[i]:.3f}     | {C1_disc[i]:.3f}")

# Plotting for MLCC and disc ceramic capacitors
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(f_mlcc, C1_mlcc, marker='o', label='MLCC')
plt.title('C1 vs Frequency for MLCC')
plt.xlabel('Frequency (kHz)')
plt.ylabel('C1 (pF)')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(f_disc, R4_disc, marker='o', label='Disc Ceramic', color='red')
plt.title('R4 vs Frequency for Disc Ceramic Capacitors')
plt.xlabel('Frequency (kHz)')
plt.ylabel('C1 (pF)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Temperature dependent data for BaTiO3 at four operating frequencies
T = [50,60,70,80,90,100,110,120,130,140,150] # °C
R4_15 = [0.888,0.900,0.950,0.986,1,1.106,1.230,1.434,1.704,1.746,1.556] # kOhm at 15 kHz
R4_25 = [0.882,0.892,0.940,0.976,1,1.094,1.210,1.416,1.652,1.702,1.512] # kOhm at 25 kHz
R4_35 = [0.878,0.888,0.934,0.974,1,1.088,1.196,1.404,1.606,1.658,1.478] # kOhm at 35 kHz
R4_45 = [0.876,0.882,0.928,0.968,1,1.088,1.192,1.400,1.604,1.656,1.460] # kOhm at 45 kHz

C1_15 = [ (1000*R4_15[i])/3 for i in range(len(R4_15))] # pF at 15 kHz
C1_25 = [ (1000*R4_25[i])/3 for i in range(len(R4_25))] # pF at 25 kHz
C1_35 = [ (1000*R4_35[i])/3 for i in range(len(R4_35))] # pF at 35 kHz
C1_45 = [ (1000*R4_45[i])/3 for i in range(len(R4_45))] # pF at 45 kHz

e_15 = [C1_15[i]/C0 for i in range(len(R4_15))] # dimensionless at 15 kHz
e_25 = [C1_25[i]/C0 for i in range(len(R4_25))] # dimensionless at 25 kHz
e_35 = [C1_35[i]/C0 for i in range(len(R4_35))] # dimensionless at 35 kHz
e_45 = [C1_45[i]/C0 for i in range(len(R4_45))] # dimensionless at 45 kHz

# plotting temperature dependence
plt.figure(figsize=(12, 6))
plt.plot(T, e_15, marker='<', linestyle='--', label='15 kHz')
plt.plot(T, e_25, marker='o', linestyle='-', label='25 kHz')
plt.plot(T, e_35, marker='s', linestyle='-.', label='35 kHz')
plt.plot(T, e_45, marker='^', label='45 kHz')
plt.title('Dielectric Constant (e) vs Temperature for BaTiO3')
plt.xlabel('Temperature (°C)')
plt.ylabel('Dielectric Constant (e)')
plt.legend()
plt.show()

```

```python
# Diffused parameter (delta) calculation
T_c = 140 # °C
e_max = [max(e_15), max(e_25), max(e_35), max(e_45)] # maximum dielectric constant at each frequency
# ln(1/e - 1/e_max) vs ln(T - T_c) for each frequency
diff_e_15 = [np.log(1/e_15[i] - 1/e_max[0]) for i in range(10,len(e_15))]
diff_e_25 = [np.log(1/e_25[i] - 1/e_max[1]) for i in range(10,len(e_25))]
diff_e_35 = [np.log(1/e_35[i] - 1/e_max[2]) for i in range(10,len(e_35))]
diff_e_45 = [np.log(1/e_45[i] - 1/e_max[3]) for i in range(10,len(e_45))]
ln_TT_c = [np.log(T[i] - T_c) for i in range(10,len(T))] 

# plotting diffused parameter
plt.figure(figsize=(12, 6)) 
plt.plot(ln_TT_c, diff_e_15, marker='<', linestyle='--', label='15 kHz')
plt.plot(ln_TT_c, diff_e_25, marker='o', linestyle='-', label='25 kHz')
plt.plot(ln_TT_c, diff_e_35, marker='s', linestyle='-.', label='35 kHz')
plt.plot(ln_TT_c, diff_e_45, marker='^', label='45 kHz')
plt.title('Diffused Parameter (delta) Calculation for BaTiO3')      
plt.xlabel('ln(T - T_c)')
plt.ylabel('ln(1/e - 1/e_max)')
plt.legend()
plt.show()
```
